```python
import os
import sys

names: ['Apple Scab Leaf', 'Apple leaf', 'Apple rust leaf', 'Bell_pepper leaf spot', 'Bell_pepper leaf', 'Blueberry leaf', 'Cherry leaf', 'Corn Gray leaf spot', 'Corn leaf blight', 'Corn rust leaf', 'Peach leaf', 'Potato leaf early blight', 'Potato leaf late blight', 'Potato leaf', 'Raspberry leaf', 'Soyabean leaf', 'Soybean leaf', 'Squash Powdery mildew leaf', 'Strawberry leaf', 'Tomato Early blight leaf', 'Tomato Septoria leaf spot', 'Tomato leaf bacterial spot', 'Tomato leaf late blight', 'Tomato leaf mosaic virus', 'Tomato leaf yellow virus', 'Tomato leaf', 'Tomato mold leaf', 'Tomato two spotted spider mites leaf', 'grape leaf black rot', 'grape leaf']

# transform to normal/disease
# 3개 이상 단어로 구성된 label은 disease로 분류
# 2개 단어로 구성된 label은 normal로 분류
# index 를 저장

disease_list = []
normal_list = []

for i in range(len(names)):
    if len(names[i].split()) >= 3:
        disease_list.append(str(i))
    else:
        normal_list.append(str(i))

print(disease_list)
print(normal_list)
print(len(normal_list))
```

```python
# disease_list에 들어가는 애들은 전부 0으로 바꿔
# 나머지는 1로 바꿈

ex_file = os.listdir('C:/Users/sabin/Desktop/dataset/PlantDoc-Object-detection/train/labels')
# open ex_file[0]

non_readable = []
for i in range(len(ex_file)):
    try:    
        f = open("C:/Users/sabin/Desktop/dataset/PlantDoc-Object-detection/train/labels/"+ex_file[i], 'r', encoding='utf-8')
        lines = f.readlines()
        f.close()
    except:
        non_readable.append(ex_file[i])
        continue
    for j in range(len(lines)):
        splited_line = lines[j].split()
        if splited_line[0] in disease_list:
            print("disease")
            lines[j] = '0 ' + splited_line[1] + ' ' + splited_line[2] + ' ' + splited_line[3] + ' ' + splited_line[4] + '\n'
        else:
            print("normal")
            lines[j] = '1 ' + splited_line[1] + ' ' + splited_line[2] + ' ' + splited_line[3] + ' ' + splited_line[4] + '\n'
    f = open("C:/Users/sabin/Desktop/dataset/PlantDoc-Object-detection/train/labels/"+ex_file[i], 'w', encoding='utf-8')
    f.writelines(lines)
    f.close()
```
