```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

prop_class_1 = np.arange(0, 1.001, 0.001)
prop_class_2 = 1 - prop_class_1

classification_error = 1 - np.maximum(prop_class_1, prop_class_2)

gini = prop_class_1*(1-prop_class_1) + prop_class_2*(1-prop_class_2)

entropy = -prop_class_1*np.log(prop_class_1) - prop_class_2*np.log(prop_class_2)

data = pd.DataFrame({'prop_class_1': prop_class_1, 'prop_class_2': prop_class_2, 
                     'classification_error': classification_error, 'gini': gini, 'entropy': entropy})
data = pd.melt(data, id_vars=['prop_class_1', 'prop_class_2'], 
               value_vars=['classification_error', 'gini', 'entropy'], var_name='metric')

plt.figure(figsize=(10, 6))
plt.plot(data[data['metric'] == 'classification_error']['prop_class_1'], data[data['metric'] == 'classification_error']['value'], label='Classification Error')
plt.plot(data[data['metric'] == 'gini']['prop_class_1'], data[data['metric'] == 'gini']['value'], label='Gini')
plt.plot(data[data['metric'] == 'entropy']['prop_class_1'], data[data['metric'] == 'entropy']['value'], label='Entropy')
plt.xlabel('Proportion (of class \'1\')')
plt.ylabel('Value')
plt.title('Metrics by Proportion of Class 1')
plt.legend()
plt.show()
```
