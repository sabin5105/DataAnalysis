```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

pd.set_option('display.precision', 2)
pd.set_option('display.max_row', 10)
pd.set_option('display.float_format', '{:20,.2f}'.format)
plt.style.use('seaborn-v0_8')
```

```python
# boston housing dataset
boston = pd.read_csv('./data/Boston.csv')
boston.head()
```

```python
from sklearn.linear_model import LinearRegression

ols_s1 = LinearRegression(fit_intercept=True)

x_train = boston['lstat'].values.reshape(-1,1)
y_true = boston['medv']
ols_s1.fit(x_train, y_true)

ols_s1.intercept_, ols_s1.coef_
```

```python
from sklearn.metrics import mean_squared_error, explained_variance_score, r2_score

y_pred = ols_s1.predict(x_train)

ols_s1_summary = {'R2': r2_score(y_true, y_pred),
                  'Ex. var': explained_variance_score(y_true, y_pred),
                  'MSE': mean_squared_error(y_true, y_pred)}

for k, v in ols_s1_summary.items():
    print(f"{k}: {v}")
```

R2 = 1 - (SSR / SST) 
* SSR = sum((y_true - y_pred)²) ; SST = sum((y_true - y_true.mean())²)
  
explained_variance_score = 1 - (MSE / variance(y_true))

### R2 == explained_variance_score means that the model explains all the variance in the data

```python
# out of sample prediction
ols_s1.predict(np.array([5,10,15]).reshape(-1,1))
```

```python
import statsmodels.api as sm

x_train = boston['lstat']
y_true = boston['medv']

ols_sm = sm.OLS(y_true, sm.add_constant(x_train)).fit()
ols_sm.summary()
```

```python
ols_sm.predict(sm.add_constant(np.array([5,10,15])))
```

```python
# carseats dataset
carseats = pd.read_csv('./data/Carseats.csv', index_col=0)
carseats.head()
```

```python
def ols_formula(df, dependent_var, *excluded_cols):
    '''
    generates the R style formula for statsmodels (patsy) given
    the dataframe, dependent variable and excluded columns
    as strings
    '''
    df_columns = list(df.columns.values)
    df_columns.remove(dependent_var)
    for col in excluded_cols:
        df_columns.remove(col)
    return dependent_var + ' ~ ' + ' + '.join(df_columns)

ols_smf = smf.ols(formula=ols_formula(boston, 'medv'), data=boston).fit()

ols_smf.summary()
```

```python
form = ols_formula(carseats, 'Sales', 'Urban', 'US', 'ShelveLoc') + ' + Income:Advertising + Price:Age + C(ShelveLoc)'
ols_smf = smf.ols(formula=form, data=carseats).fit()

ols_smf.summary()
```

```python
form = ols_formula(carseats, 'Sales') + ' + Income:Advertising + Price:Age'
ols_smf = smf.ols(formula=form, data=carseats).fit()

ols_smf.summary()
```
